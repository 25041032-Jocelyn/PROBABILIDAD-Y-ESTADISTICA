```python
from statistics import *
```

```python
# Lista de calificaciones del grupo A de probabilidad y estadistica
grupoa = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(grupoa)
```

```python
# la funcion type nos devuelve el tipo de dato de la variable type
type (grupoa)
```

```python
# Sacar la media de las calificaciones con una formula
promedio = sum(grupoa)/len(grupoa)
print(promedio)
```

```python
# Utilizar la funcion mean que esta en la libreria statistics para sacar el promedio
# el promedio:
promedio2 = mean (grupoa)
print(promedio2)
```

```python
# Determinar el valor de la mediana con la funcion mediam
mediana = median(grupoa)
print (mediana)
```

```python
# Determinar el valor de la moda con la funcion mode
moda = mode(grupoa)
print(moda)
```

```python
tiempos = [10.23, 15.20, 22.15, 11.49, 17.83, 16.30, 12.53, 13.55, 10.23, 14.56]
promedio = sum(tiempos)/len(tiempos)
mediana = median(tiempos)
moda = mode(tiempos)
print ("Los tiempos de los jugadores fueron: ", tiempos)
print ("El promedio de los tiempos es: ", promedio)
print ("La mediana de los tiempos es: ", mediana)
print ("La moda de los tiempos es: ", moda)
```

```python
#libreria para genarar numeros aleatorios
import random
fabrica1 = []
for i in range (100):
  numero_aleatorio = round( random.uniform(50, 100), 2)
  fabrica1.append(numero_aleatorio)
print (fabrica1)
```

```python
fabrica2 = []
for i in range (100):
  numero_aleatorio = round( random.uniform(70, 100), 2)
  fabrica2.append(numero_aleatorio)
print (fabrica2)
```

```python
#libreria de python para graficar variables
import matplotlib.pyplot as plt
plt.hist(fabrica1)
plt.title("Poducción diaria de la fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Poducción diaria de la fabrica 2")
```

```python
media1 = mean (fabrica1)
mediana1 = median (fabrica1)
moda1 = mode (fabrica1)
print ("La media de la fabrica 1 es: ", media1)
print ("La mediana de la fabrica 1 es: ", mediana1)
print ("La moda de la fabrica 1 es: ", moda1)
```

```python
media2 = mean (fabrica2)
mediana2 = median (fabrica2)
moda2 = mode (fabrica2)
print ("La media de la fabrica 2 es: ", media2)
print ("La mediana de la fabrica 2 es: ", mediana2)
print ("La moda de la fabrica 2 es: ", moda2)
```

```python
#generar 2 sudplod de 1 renglon por 2 columnas para desplegar los 2 histogramas, una al lado de la otra
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (10, 5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python
counts1, bins1, patches1 = ax1.hist(fabrica1)
print ("Limites de Bins Fábrica 1: ", bins1)
print ("Freceuncia de la Fábrica 1: ", counts1)
```

```python
rangos = [50, 55, 60, 70, 75, 80, 85, 90, 95, 100] # definir rangos de bins
plt.hist(fabrica1, bins=rangos, alpha=0.2, color='b', edgecolor='black')
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label="Mediana")
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label="Moda")
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Distribución de la Fabrica 1')
plt.legend()
plt.show()
```
