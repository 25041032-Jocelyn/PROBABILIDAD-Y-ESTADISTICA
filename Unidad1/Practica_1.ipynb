```python
from statistics import *
```

```python
# Lista de calificaciones del grupo A de probabilidad y estadistica
grupoa = [70, 70, 65, 90, 100, 95, 45, 80, 73, 20, 100, 96]
print(grupoa)
```

```python
# la funcion type nos devuelve el tipo de dato de la variable type
type (grupoa)
```

```python
# Sacar la media de las calificaciones con una formula
promedio = sum(grupoa)/len(grupoa)
print(promedio)
```

```python
# Utilizar la funcion mean que esta en la libreria statistics para sacar el promedio
# el promedio:
promedio2 = mean (grupoa)
print(promedio2)
```

```python
# Determinar el valor de la mediana con la funcion mediam
mediana = median(grupoa)
print (mediana)
```

```python
# Determinar el valor de la moda con la funcion mode
moda = mode(grupoa)
print(moda)
```

```python
tiempos = [10.23, 15.20, 22.15, 11.49, 17.83, 16.30, 12.53, 13.55, 10.23, 14.56]
promedio = sum(tiempos)/len(tiempos)
mediana = median(tiempos)
moda = mode(tiempos)
print ("Los tiempos de los jugadores fueron: "), (tiempos)
print ("El promedio de los tiempos es: "), (promedio)
print ("La mediana de los tiempos es: "), (mediana)
print ("La moda de los tiempos es: "), (moda)
```

```python
#libreria para genarar numeros aleatorios
import random
fabrica1 = []
for i in range (100):
  numero_aleatorio = round( random.uniform(50, 100), 2)
  fabrica1.append(numero_aleatorio)
print (fabrica1)
```
